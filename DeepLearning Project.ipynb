```python
import os
import cv2
import numpy as np

X_train = []
Y_train = []

def getID(name):  # Dummy function, replace with actual logic
    return 1 if "dog" in name.lower() else 0

path = "dataset"  # Set your dataset path

for root, dirs, directory in os.walk(path):
    for j in range(len(directory)):
        name = os.path.basename(root)
        img_path = os.path.join(root, directory[j])

        # Print the current file being processed
        print(name, img_path)

        # Skip non-image files
        if not directory[j].lower().endswith(('.png', '.jpg', '.jpeg')):
            continue

        # Read the image
        img = cv2.imread(img_path)

        # Ensure image is loaded properly
        if img is None:
            print(f"Warning: Could not read image {img_path}, skipping...")
            continue

        # Resize the image
        img = cv2.resize(img, (64, 64))

        # Convert to numpy array and reshape
        im2arr = np.array(img).reshape(64, 64, 3)

        # Append to training data
        X_train.append(im2arr)
        Y_train.append(getID(name))

# Convert lists to NumPy arrays
X_train = np.asarray(X_train)
Y_train = np.asarray(Y_train)

print("Y_train:", Y_train)
```
